```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "../ETL Project/metal_bands_2017.csv"
metal_df = pd.read_csv(csv_file,encoding="ISO-8859-1")
metal_df.head()
```

```python
csv_file = "../ETL Project/WHR2017.csv"
happy_df = pd.read_csv(csv_file,encoding="ISO-8859-1")
happy_df.head()
```

```python
metal_renamed = metal_df.rename(columns={"origin": "Country"})
del metal_renamed["style"]
metal_renamed.head()
```

```python
happy_df['Country'].unique()
```

```python
happy_df['Country'] = happy_df['Country'].replace(['United States'],'USA')
happy_df['Country'].unique()
```

```python
# remove or move 2nd country names from metal lists
# pivot on country name
```

```python
# created additional column to house second country listed
metal_renamed['C1'], metal_renamed['C2'] = metal_renamed['Country'].str.split(',', 1).str
metal_renamed['C1'].unique()
```

```python
# check columns
metal_renamed.head()
```

```python
del metal_renamed["Country"]
metal_renamed.head()
```

```python
del metal_renamed["C2"]
metal_renamed.head()
```

```python
# rename C1 to Country
metal_renamed = metal_renamed.rename(columns={"C1": "Country"})
metal_renamed.head()
```

```python
del metal_renamed["Unnamed: 0"]
metal_renamed.head()
```

```python
merged_df = pd.merge(metal_renamed, happy_df, on="Country")
merged_df.head()
```

```python
# rename all columns to nice things
merged_df = merged_df.rename(columns={
                            "band_name": "Band",
                            "fans": "Fans",
                            "formed": "Formed",
                            "split": "Consciously Uncoupled",
                            "Happiness.Rank": "Happiness Rank",
                            "Happiness.Score": "Happiness Score",
                            "Whisker.high": "Whisker High",
                            "Whisker.low": "Whisker Low",
                            "Economy..GDP.per.Capita.": "GDP Per Capita",
                            "Health..Life.Expectancy.": "Life Expectancy",
                            "Trust..Government.Corruption.": "Trust/Corruption",
                            "Dystopia.Residual": "Dystopia Index"})
merged_df.head()
```

```python
connection_string = "postgres:packers@localhost:5432/MetalHappy"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
happy_df.to_sql(name='Happy', con=engine, if_exists='append', index=True)
```

```python
metal_renamed.to_sql(name='Metal', con=engine, if_exists='append', index=True)
```

```python
merged_df.to_sql(name='Merged', con=engine, if_exists='append', index=True)
```
